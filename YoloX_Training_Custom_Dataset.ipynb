```python
import os
import random

trainval_percent = 0.1
train_percent = 0.9
xmlfilepath = '/content/gdrive/Othercomputers/My Laptop/Repositories/YOLOX_Training_Custom_DataSet/YOLOX/datasets/VOCdevkit/VOC_BCCD/Annotations'
txtsavepath = '/content/gdrive/Othercomputers/My Laptop/Repositories/YOLOX_Training_Custom_DataSet/YOLOX/datasets/VOCdevkit/VOC_BCCD/ImageSets'
total_xml = os.listdir(xmlfilepath)

num = len(total_xml)
list = range(num)
tv = int(num * trainval_percent)
tr = int(tv * train_percent)
trainval = random.sample(list, tv)
train = random.sample(trainval, tr)

ftest = open(txtsavepath+'/Main/test.txt', 'w')
ftrain = open(txtsavepath+'/Main/trainval.txt', 'w')

for i in list:
    name = total_xml[i][:-4] + '\n'
    if i in trainval:
        ftest.write(name)
    else:
        ftrain.write(name)

ftrain.close()
ftest.close()
```
